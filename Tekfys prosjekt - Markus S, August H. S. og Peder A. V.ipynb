#                       Damptrykkurven for $H_20$

**Innledning**

I dette prosjektet skal man se nærmere på numerisk løsning av $\text{PDE}$ (Partial Differential Equation). Vi vil primært fokusere på newtonsmetode med én og flere variabler. For å teste disse metodene skal man ta utgangspukt i ligninger fra termisk fysikk, nærmere bestemt damptrykk kurven til $H_2O$. Man skal finne denne kurven ved anlytiske og numeriske metoder, og sammenligne dette med eksperimentell data. 

**Teori**

For å endre aggregattilstand  til $H_2O$ fra flytende til gass må energi tilføres. Hvor mye energi som kreves bestemmes av trykk, volum, stoffmengde og temperatur til systemet. Sammenhengen mellom disse fire tilstandsvariblene er beskrevet av ideell gasslov:

$$
p = \frac{nRT}{V}
$$

der $n$ er antall mol, $R$ er den ideelle gasskonstanten, $T$ er temperatur og $V$ er volum. Denne ligningen tar ikke hensyn til volumet til molekylene eller intermolekylære krefter. En mer realistisk modell er Van der Waals tilstandsligning:

$$
p = \frac{nRT}{V-nb} -\frac{n^2a}{V^2},
$$

der $b$ er en konstant som tar hensyn til volumet opptatt av molekylene og $a$ er en konstant som tar hensyn til tiltrekkende krefter mellom molekylene. I faseovergangen vil Van der Waals tilstandsligning bryte sammen. I realiteten vil trykket og temperaturen være konstant i faseovergangen, mens volumet vil endre seg ettersom $H_2O$ går fra væske til gass eller motsatt. Volumet når all $H_2O$ er i væsketilstand kaller vi $V_v$, mens volumet når all $H_2O$ er i gasstilstand kaller vi for $V_g$.Trykket ved faseovergangen avhenger av temperaturen, og kan beskrives av Clapeyrons ligning.

Clapeyrons ligning er en funksjon av entropi, volum og temperatur:

$$
\frac{dp}{dT} = \frac{\Delta S}{\Delta V},
$$

der $\Delta S$ er endringen i entropi og $\Delta V$ er volumendringen, fra gass til væske fase. Entropiendringen er gitt ved overgangsvarmen fra væske til gass, dividert på temperatur. Denne varmen kalles for latent varme $L(T)$:

$$
\Delta S = \frac{L(T)}{T}.
$$

Da får man videre at uttrykket for damptrykkurven blir:

$$
\frac{dp}{dT} = \frac{L(T)}{T \Delta V(T) },
$$

der $\Delta V(T) = V_g - V_v$ er volumendringen ved en gitt temperatur.

Clapeyrons ligning er kun gyldig i området der man har direkte overgang fra væske til gass, dette vil si fra trippelpunktet til kritisk punkt for $H_2O$. Trippelpunktet til $H_2O$ er punktet der stoffet kan eksistere i alle tre tilstander, og er definert som $t = 0.01 \: ^\circ C$. Kritisk punkt er definert som punktet hvor forskjellen mellom væske og gassfase opphører, og stoffet blir et superkritisk fluid. Dette skjer ved kritisk trykk $p_c$ og kritisk temperatur $T_c$.  Disse kritiske verdiene bestemmes eksperimentelt og kan brukes til å finne konstantene $a$ og $b$, samt kritisk volum $V_c$.

Med utgangspunkt i termisk fysikk finner man følgende uttrykk for $p_c$, $T_c$ og $V_c$:

$$ 
p_c = \frac{a}{27b^2}, T_c = \frac{8a}{27Rb},  V_c = 3b.
$$

Fra disse utrykkene kan man videre finne uttrykk for konstantene $a$ og $b$:

$$
a = 27p_cb^2,\: b = \frac{8a}{27RT_c} \Rightarrow a = 27p_c\left(\frac{8a}{27RT_c}\right)^2 \Leftrightarrow  a = \frac{27R^2T_c^2}{64p_c}
$$

$$
b = \frac{8a}{27RT_c} = \frac{8}{27RT_c}\left( \frac{27R^2T_c^2}{64p_c}\right) \Leftrightarrow b = \frac{RT_c}{8p_c}
$$

Fra $[1]$ finner man at kritisk temperatur er $T_c = 647.096 \; K$ og kritisk trykk er $p_c = 220.64 \; bar$ for $H_2O$. 

Setter man inn verdiene i uttrykket for $a$ og $b$ får man:

$$
 a = \frac{27R^2T_c^2}{64p_c} =  \frac{27\cdot \left(8.314 \frac{m^3\cdot Pa}{K\cdot mol}\right)^2 \cdot (647.096\: K)^2}{64\cdot (220.64 \cdot 10^5 \:Pa)} \approx 0.5534 \; m^6Pa\:mol^{-2} = 5.534 \: L^2\:bar\:mol^{-2}
$$

$$
b = \frac{RT_c}{8p_c} = \frac{\left(8.314 \frac{m^3\cdot Pa}{K\cdot mol}\right) \cdot 647.096\; K}{8\cdot 220.64 \cdot 10^5 \;Pa } \approx 3.048 \cdot 10^{-5} \; m^3mol^{-1} = 0.03048 \;L\:mol^{-1}
$$

Det molare kritiske volumet er da:

$$
 V_c = 3b = 3 \cdot 0.03048  \;L\:mol^{-1} = 0.09144 \;L\:mol^{-1}
$$

Eksperimentelle målinger fra $[2]$ og molar masse fra $[3]$, viser at det molare kritiske volumet til $H_2O$ er:

$$
V_c = 3.10559 cm^3/g \cdot (16.00 + 2\cdot 1.008)g/mol= 55.95 cm^3/mol = 0.05595\; L\:mol^{-1}
$$

Man ser da et stort avvik mellom eksperimentelle målinger og kritisk volum fra Van der Waals tilstandsligning. Det absolutte avviket er på $0.03549\; Lmol^{-1}$ og relativt avvik på $63.4\%$. Avviket viser til en svakhet i Van der Waals tilstandsligning, og tydeliggjør at ligningen bryter sammen når man nærmer seg overgangsverdier. 


Nedenfor bruker man Van der Waals tilstandsligning til å plotte et $pV$-diagram med konstant temperatur $T = T_c = 647.096 K$.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as linalg
import scipy
import scipy.integrate as integrate
import scipy.interpolate as interpolate 
```

```python
#Definerer konstanter 
R = 8.314            #J/K*mol
T_c = 647.096        #K
p_c = 22.064*10**6   #Pa
n = 1                #mol

a = 27*R**2*T_c**2/(64*p_c) #Pa m^6/mol^2
b = R*T_c/(8*p_c)           #m^3/mol
```

```python
#1b)
#Velger volum fra 60 mL istedet for 75 mL som oppgitt for å få et bedre plott

#Finner koeksistenstrykket
V = np.linspace(0.060, 0.300, 1000)*1e-3  #m^3
p = n*R*T_c/(V -n*b) - (a*n**2)/(V**2)    #Pa

#Plotter
plt.plot(V*1e6, p*1e-5)
plt.xlabel('V [ml]')
plt.ylabel('p [bar]')
plt.title(fr"Trykk ved $T_c$ = {T_c} K") 
plt.show()
```

Den kvalitative forskjellen mellom grafen ovenfor og $\text{figur 2}$ i $[4]$ er i faseovergangen mellom væskefasen og gassfasen. I $\text{figur 2}$ ser vi et tredjegradspolynom med bunnpunkt og toppunkt. Dette bunn- og toppunktet befinner seg begge i overgangen mellom gass og væske fase. Dette er et ufysisk område, siden trykket i realiteten vil være konstant. I figuren over ser man et tredjegradspolynom med et terassepunkt. Isotermen ligger på den kritiske temperaturen $T_c$ og man vil dermed ikke ha noen faseovergang. Terassepunktet vil da ligge på det eksperimentellt kritiske trykket $p_c$ og den analytisk beregnede verdien for $V_c$. 

For å kunne benytte Van der Waals ligning må man unngå det ufysiske området i faseovergangen. Dette kan gjøres ved å finne startvolum ($V_v$) og sluttvolum ($V_g$). En mulig måte å gjøre dette på er newton's metode.

Newton's metode er en numerisk løsningsmetode for ligningsett av førsteordens differensialligninger. Man setter opp ligningene slik at høyresiden blir $0$. Man finner så når tangenten til funksjonen i startpunktet $x_0$ krysser $x$-aksen. Dette punktet blir da $x_1$, og man repeterer forrige steg. På denne måten itererer man seg nærmere nullpunktet for funksjonen. Newton's metode vil ikke alltid konvergere mot en løsning avhengig av startverdi og funksjonen. Matematisk for én variabel, har man:

$$
f'(x_0)(x-x_0) = f(x)-f(x_0), \; f(x) = 0
$$

$$
\Rightarrow f'(x_0)(x-x_0) = -f(x_0) 
$$

$$
\Rightarrow (x-x_0) = -\frac{f(x_0)}{f'(x_0)}
$$

$$
\Rightarrow x = x_0 -\frac{f(x_0)}{f'(x_0)}
$$

Som kan skrives på generell form: 

$$
x_{k+1} = x_k − \frac{f(x_k)}{f'(x_k)}
$$


Løsningsmetoden blir essensielt den samme for to variabler, men vil løses som et ligningssett. Dette ligningssettet kan skrives på kompakt form som:

$$
f(\hat{x}) + J(\hat{x})(x-\hat{x}) = 0,
$$

der $J(\hat{x})$ er Jacobimatrisa gitt av

$$
J(\hat{x}) = \begin{pmatrix} \frac{\partial f_1}{\partial x_1}(x_1,x_2) & \frac{\partial f_1}{\partial x_2}(x_1,x_2) \\
      \frac{\partial f_2}{\partial x_1}(x_1,x_2) & \frac{\partial f_2}{\partial x_2}(x_1,x_2)
    \end{pmatrix}
$$

En mer dataljert utledning finner man i $[4]$.

Man vil nå bruke newton's metode for å løse en test ligning, med analytisk løsning. Denne ligningen er:

$$
\sinh^2 \left(\frac{2c}{T_c}\right) -1 = 0.
$$

Den deriverte med hensyn på $T_c$ vil da ha uttrykket:

$$
\frac{d}{dT_c}\left(\sinh^2 \left(\frac{2c}{T_c}\right) -1\right) = -\frac{4c\cosh\left(\frac{2c}{T_c}\right)\sinh\left(\frac{2c}{T_c}\right)}{T_c^2}.
$$

Da får man punktene:

$$
T_{k+1} = T_k + \frac{ T_k^2 (\sinh^2 \left(\frac{2c}{T_k}\right) -1)}{4c\cosh\left(\frac{2c}{T_k}\right)\sinh\left(\frac{2c}{T_k}\right)}.
$$


```python
#1c)
c = 1 #K

def NewtonsMethod(maxIter, T_0, atol):
    T = np.zeros(maxIter)                   #Lager array
    T[0] = T_0                              #Setter startverdi
    T_c = 2*c/(np.log(1+np.sqrt(2)))        #Analytisk løsning
    count = 0                               #Teller opp antallet iterasjoner
    #Løser med Newton's metode
    for i in range(maxIter-1):
        T[i+1] = T[i] - (np.sinh(2*c/T[i])**2-1)/((-4*c*np.cosh(2*c/T[i])*np.sinh(2*c/T[i]))/T[i]**2)
        count +=1
        #Avslutter når toleransen er oppnådd
        if (np.abs(T_c-T[i+1]) < atol):
            break
            
    # Returnerer antallet iterasjoner, Temperatur og array for plotting 
    return count, T, np.linspace(0,maxIter-1,maxIter)

count1, T1, i1 = NewtonsMethod(10, 1, 1e-10)
count2, T2, i2 = NewtonsMethod(10, 3, 1e-10)
count3, T3, i3 = NewtonsMethod(10, 4, 1e-10)
count3, T3, i3 = NewtonsMethod(10, 4, 1e-10)
count4, T4, i4 = NewtonsMethod(10, 2, 1e-10)
T_c = 2*c/(np.log(1+np.sqrt(2)))

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))

#Plott nr 1
plt.subplot(2, 2, 1)
plt.plot(i1[:count1+1], T1[:count1+1], color = "purple", label =  r"$T_i$ numerisk")
plt.axhline(T_c, color = "red", label = r"$T_c$ analytisk")
plt.xlabel("iterasjoner")
plt.ylabel(r"$T_c$")
plt.title(fr"Kritisk temperatur, $T_c$, $T_0 =$  {T1[0]} ")
plt.legend()

#Plott nr 2
plt.subplot(2, 2, 3)
plt.plot(i2[:count2+1], T2[:count2+1], color = "purple", label =  r"$T_i$ numerisk")
plt.axhline(T_c, color = "red", label = r"$T_c$ analytisk")
plt.xlabel("iterasjoner")
plt.ylabel(r"$T_c$")
plt.title(fr"Kritisk temperatur, $T_c$, $T_0 =$  {T2[0]} ")
plt.legend()

#Plott nr 3
plt.subplot(2, 2, 4)
plt.plot(i3[:count3+1], T3[:count3+1], color = "purple", label =  r"$T_i$ numerisk")
plt.axhline(T_c, color = "red", label = r"$T_c$ analytisk")
plt.xlabel("iterasjoner")
plt.ylabel(r"$T_c$")
plt.title(fr"Kritisk temperatur, $T_c$, $T_0 =$  {T3[0]} ")
plt.legend()

#Plott nr 4
plt.subplot(2, 2, 2)
plt.plot(i4[:count4+1], T4[:count4+1], color = "purple", label =  r"$T_i$ numerisk")
plt.axhline(T_c, color = "red", label = r"$T_c$ analytisk")
plt.xlabel("iterasjoner")
plt.ylabel(r"$T_c$")
plt.title(fr"Kritisk temperatur, $T_c$, $T_0 =$  {T4[0]} ")
plt.legend()
plt.show()
```

Man observerer at newton's metode konvergerer ganske raskt uavhengig av startverdi. Derimot varierer verdien den konvergerer mot. Fra plottet nederst til høyre observerer man at temperaturen konvergerer mot den negative verdien av $T_c$, som ikke gir fysisk mening. Hvor mange iterasjoner man må ha varierer ut fra startpunkt $T_0$ og man observerer at startpunkt nær $T_c$ vil ofte behøve færre iterasjoner. 

Fra ligning $(16)$ i $[4]$ ser man følgende formel:

$$
\lim_{i\rightarrow \infty} \frac{|x_{i+1}-r|}{|x_i-r|^q} = \mu,
$$

der $\mu$ er konvergensraten, $r$ er roten av funksjonen $f(x)$ og $q$ er konvergensordenen. For å si noe om hvor effektivt metoden konvergerer må vi finne konvergensordenen. Konvergensordenen deles inn i tre kategorier, sublineær-, lineær- og superlinær konvergens. Ved sublineær konvergens er $q<1$, lineær konvergens er $q = 1$ og superlineær konvergens er $q>1$. Det er vanskelig å bruke denne formelen direkte, men det kan vises at følgen under konvergerer mot $q$:

$$
p_i = \frac{\log\: (e_i/e_{i-1})}{\log\: (e_{i-1}-e_{i-2})}. 
$$

Newton's metoden har som kjent kvadratisk konvergens, altså $q=2$. Dette skal man vise i kodeblokken under.

```python
#1d)

#Definerer en e array som viser hvordan den absolutte feilien endrer seg med hver iterasjon.
e = np.abs(T1 - T_c)

#Kalkulerer følgen
p_i = np.zeros(len(e)-1)
for i in range(2,len(e)-1):
    p_i[i]= np.log(e[i]/e[i-1])/(np.log(e[i-1]/e[i-2]))


fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

#Plotter error 
plt.subplot(1, 2, 1)
plt.plot(i1[:count1+1], e[:count1+1], color = "purple")
plt.xlabel("i")
plt.ylabel(r"$e_i$")
plt.title(fr"Error $e_i$, for $i$ iterasjoner  og $T_0 =$  {T1[0]} ")

#Plott følgen p_i
plt.subplot(1, 2, 2)
plt.plot(i1[:count1+1], p_i, color = "purple")
plt.xlabel("i")
plt.ylabel(r"$p_i$")
plt.axhline(y = 2, linestyle = 'dashed', color = 'black')
plt.title(fr" $p_i$, for $i$ iterasjoner  og $T_0 =$  {T1[0]} ")
plt.show()
```

Ser på plottet for $p_i$ at følgen konverderer mot $2$, altså kan man fastslå at newtons metode har kvadratisk kovergens. Dette stemmer overens med absolutt feil $e_i$ som tydelig fra plottet har superlineær konvergens.

Til nå har man sett på newton's metode for en variabel, men videre skal man utvide metoden til to variabler. For å teste dette skal man gå tilbake til damptrykkurven. Vi vil nærmere bestemt finne et numerisk uttrykk for væskevolumet $V_v$ og gassvolumet $V_g$, for ett mol $H_2O$.

Som tidligere diskutert vil man i realiteten ha konstant trykk i faseovergangen fra væske til gass. Man kan dermed sette at trykket ved væskevolumet $V_v$ tilsvarer trykket ved gassvolumet $V_g$, som beskrives av ligningen under:

$$
p_0 = \frac{RT}{V_g-b}-\frac{a}{V_g^2} = \frac{RT}{V_v-b}-\frac{a}{V_v^2}.
$$

Hvis man kombinerer dette med maxwell's konstruksjon finner man følgende ligning:

$$
\frac{RT}{V_g − V_v} ln \left( \frac{V_g − b}{V_v − b} \right) − \frac{a}{V_gV_v} = \frac{RT}{V_g − b}− \frac{a}{V_g^2}.
$$

Ved å omskrive disse ligningene til funksjoner $f(V_v,V_g) = 0$ og $g(V_v,V_g) = 0$ får man:

$$
f(V_v,V_g) = \frac{RT}{V_g-V_v}ln\left(\frac{V_g-b}{V_v-b}\right) -\frac{a}{V_gV_v}-\frac{RT}{V_g-b}+\frac{a}{V_g^2} = 0
$$

$$
g(V_v,V_g) = \frac{RT}{V_g-b}-\frac{a}{V_g^2} - \frac{RT}{V_v-b}+ \frac{a}{V_v^2} = 0
$$

Man har nå to ligninger og to variabler, $V_v$, $V_g$, og kan bruke newton's metode til å finne verdien de konvergerer mot. Vi starter med å bestemme Jacobimatrisen $J(V_v,V_g)$ for ligningsystemet:

$$
\frac{\partial f}{\partial V_g} = -\frac{RT}{(V_g-V_v)^2}ln\left(\frac{V_g-b}{V_v-b}\right)+\frac{RT}{(V_g-b)(V_g-V_v)}+\frac{a}{V_g^2V_v}+\frac{RT}{(V_g-b)^2} - \frac{2a}{V_g^3} 
$$

$$
\frac{\partial f}{\partial V_v} = \frac{RT}{(V_g-V_v)^2}ln\left(\frac{V_g-b}{V_v-b}\right)-\frac{RT}{(V_v-b)(V_g-V_v)}+\frac{a}{V_gV_v^2} 
$$

$$
\frac{\partial g}{\partial V_g} = \frac{2a}{V_g^3}-\frac{RT}{(V_g-b)^2}
$$

$$
\frac{\partial g}{\partial V_v} = \frac{RT}{(V_v-b)^2} - \frac{2a}{V_v^3}
$$

Jacobimatrisen blir da:

$$
J(V_v,V_g) = \begin{pmatrix} \frac{\partial f}{\partial V_v} & \frac{\partial f}{\partial V_g} \\
\frac{\partial g}{\partial V_v} & \frac{\partial g}{\partial V_g} \end{pmatrix}
$$

Nå kan vi kombinere alt til newton's metode for to variabler.


```python
#Setter max antall iterasjoner og toleranse
maxIter = 50
atol = 1e-8
```

```python
#Definerer funksjonene våre
def func(T, V_v, V_g):
    f = ((R*T)/(V_g-V_v)) * np.log((V_g-b)/(V_v-b))-a/(V_g*V_v)-(R*T)/(V_g-b) + a/V_g**2
    g = (R*T)/(V_g-b) - a/V_g**2 - (R*T)/(V_v-b) + a/V_v**2
    return np.array([f, g])

#Definerer jacobimatrisen
def jacobi(T, V_v, V_g):    
    df_Vg = -((R*T)/(V_g-V_v)**2)*np.log((V_g-b)/(V_v-b))+ (R*T)/((V_g-b)*(V_g-V_v))+ a/(V_g**2*V_v)+(R*T)/(V_g-b)**2-(2*a)/V_g**3
    df_Vv = ((R*T)/(V_g-V_v)**2)*np.log((V_g-b)/(V_v-b)) - (R*T)/((V_v-b)*(V_g-V_v)) + a/(V_v**2*V_g)

    dg_Vg = (2*a)/V_g**3 -(R*T)/(V_g-b)**2
    dg_Vv = (R*T)/(V_v-b)**2 -(2*a)/V_v**3
    
    return np.array([[df_Vv, df_Vg],
                     [dg_Vv, dg_Vg]])
    
#Definerer newton's metode
def newton2D(temp, V_v0, V_g0):
    #Setter startverdier
    V_v = V_v0
    V_g = V_g0
    
    #Iterer gjennom for å predikere gass og væske volum
    for i in range(maxIter):
        f = func(temp, V_v, V_g)
        J = jacobi(temp, V_v, V_g)
        #Løser for delta
        delta = linalg.solve(J, -f)
        #Oppdaterer verdiene våre
        V_v += delta[0]
        V_g += delta[1]
        
        #Sjekker når endringen for hver iterasjon kommer under toleransen, og bryter løkken
        if(linalg.norm(func(temp, V_v, V_g)) < atol):
            break
    return V_v, V_g


#Setter startverdier
T = np.linspace(274,647, 647-274+1)
V_g = np.zeros(len(T))
V_v = np.zeros(len(T))
V_g0 = 12.6*1e-3
V_v0 = 35.7*1e-6

#Løser for alle de ulike temperaturene
V_v[0], V_g[0] = newton2D(T[0],V_v0, V_g0)
for i in range(1, len(T)):
    V_v[i], V_g[i] = newton2D(T[i],V_v[i-1], V_g[i-1])
    
    
#Plotter
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

#Plotter vanlig  
plt.subplot(1, 2, 1)
plt.plot(T, V_v*1000, label = fr'væske ($V_v$)')
plt.plot(T, V_g*1000, label = fr'gass ($V_g$)')
plt.title(r"$V_g$ og $V_v$ for $H_2O$, ved ulike temperaturer $T$")
plt.xlabel("T [K]")
plt.ylabel(r"V [L]")
plt.legend()

#Plotter logaritmisk 
plt.subplot(1, 2, 2)
plt.semilogy(T, V_v*1000, label = fr'væske ($V_v$)')
plt.semilogy(T, V_g*1000, label = fr'gass ($V_g$)')
plt.title(r"$V_g$ og $V_v$ for $H_2O$, ved ulike temperaturer $T$")
plt.xlabel("T [K]")
plt.ylabel(r"$log\:(V)$")
plt.legend()
plt.show()
```

Vi observerer fra plottene at gassvolumet vil minke ved økende temperatur, mens væskevolumet vil øke. Man observer også at endringen i gassvolum fra trippelpunkt til kritisk punkt er mye større enn endringen i væskevolum. $\Delta V$ vil være størst ved lavere temperaturer og minker når temperaturen øker. Til sist observerer man at volumene konvergerer mot samme verdi analytisk beregnet verdi $V_c$. Dette gir mening da man ikke har noen faseovergang ved kritisk temperatur.

Nedenfor har vi hentet nødvendige eksperimentelle verdier fra $[5]$ og $[6]$.

```python
#Lagrer de eksperimentelle verdiene for latent varme, volum til væske og volum til gass i numpy arrayer 

#molvekt for vann, [g/mol]
molvekt_vann = 18.016 

# latent varme per mol array, [Joule/mol]
latent_varme =np.array([45054,44970,44883,44627,44456,44287,44200,43988,43774,43602,43345,43172,42911,42738,42475,42030,
41579,41120,40839,40650,40167,39671,38630,37508,36286,34944,33462,31804,29934,27798,25304,22310,18507,12967,0]) 

# temperatur til latent varme [K]
temp_latent_varme = np.array([0.01,2,4,10,14,18,20,25,30,34,40,44,50,54,60,70,80,90,96,100,110,120,140,160,180,200,220,
240,260,280,300,320,340,360,373.946]) + 273 

#tempteratur for volum til gass og væske, [K]
temp_volum=np.array([273.16,280,300,320,340,360,390,400,420,440,460,480,500,530,560,590,620,630,640,647.1]) 

#tetthet til vann,  [kg/m^3]
tetthet_væske = np.array([999.8,999.9,996.5,989.4,979.5,967.4,953.3,937.5,919.9,900.7,879.6,856.5,831.3,788.5,737.8,674.8,
                          586.9,544.2,481.5,322.0]) 

#omgjør fra tetthet, til volum per mol, [m^3/mol]
volum_væske = molvekt_vann*(tetthet_væske**-1)*10**(-3)

#tetthet til gass, [kg/m^3]
tetthet_gass = np.array([0.00485,0.00767,0.02558,0.07166,0.1744,0.3786,0.7483,1.369,2.352,3.833,5.983,9.014,13.20,22.47,
                         37.15,61.24,106.3,132.8,177.1,322.0])

#omgjør fra tetthet, til volum per mol, [m^3/mol]
volum_gass = molvekt_vann*10**(-3)/tetthet_gass
```

Ut i fra plottene over ønsker vi å finne tilnærmede funksjoner for dataene. Dette gjør vi ved å optimalisere ulike funksjonsuttrykk, og sjekke hvilke som passer best med dataen. Implementeringen for dette er gjort i kodeblokkene under.

```python
#2b)
T_c = 647.096 #kritisk temperatur [K] 
def L_funk(T,a,b,c): 
    """
    latent varme som funksjon av T
    input:
        T: temperaturen i punktet 
    output: 
        latente varmen ved temperatur T 
    """
    return a * abs(T_c - T)**b + c 

def V_funk(T,a,b,c): 
    """
    volumet til væsken som funksjon av T 
    input:
        T: temperaturen i punktet 
    output:
        volumet til væsken ved temperatur T 
    """
    return a * np.abs(T_c - T)**b + c

def G_funk(T,a,b,c,d): 
    """
    volumet til gassen som funksjon av T 
    input:
        T: temperaturen i punktet 
    output:
        volumet til gassen ved temperatur T 
    """
    return a * T**3 + b * T**2 + c * T + d
```

Over har man funnet funksjonene, som passer godt til datapunktene. Dette har vi gjort ved å prøve diverse funksjoner og polynomer. Videre skal man optimalisere disse parameterene.

```python
#beregner parameterte til L_funk, V_funk og G_funk med scipy.optimize.curve_fit
param_Latent_varme, param_cov_Latent_varme = scipy.optimize.curve_fit(L_funk, temp_latent_varme, latent_varme) 
param_væske, param_cov_væske = scipy.optimize.curve_fit(V_funk, temp_volum, volum_væske)

#Her bruker vi logaritmen av volum_gass for å få penere parametere 
param_gass, param_cov_gass = scipy.optimize.curve_fit(G_funk, temp_volum,np.log(volum_gass)) 
```

Nedenfor definerer vi funksjoner med de optimaliserte parameterene.

```python
#lager ny funksjoner for latent varme og volum til væske og gass med de korrekte parameterene
def latent_varme_func(T): 
    """
    latent varme som funksjon av T med parametere beregnet fra scipy.optimize.curve_fit
    input:
        T: temperaturen i punktet 
    output: 
        latente varmen ved temperatur T 
    """
    return param_Latent_varme[0] * np.abs(T_c - T)**param_Latent_varme[1] + param_Latent_varme[2]      #[Joule]

def væske_volum_func(T):
    """
    volumet til væsken som funksjon av T med parametere beregnet fra scipy.optimize.curve_fit
    input:
        T: temperaturen i punktet 
    output:
        volumet til væsken ved temperatur T 
    """
    return param_væske[0] * np.abs(T_c - T)**param_væske[1] + param_væske[2]                             #[m^3]

def gass_volum_func(T):
    """
    volumet til gassen som funksjon av T med parametere beregnet fra scipy.optimize.curve_fit
    input:
        T: temperaturen i punktet 
    output:
        volumet til gassen ved temperatur T 
    """
    #Opphøyer i exp for å omgjøre fra logaritmisk skala
    return np.exp(param_gass[0] * T**3 + param_gass[1] * T**2 + param_gass[2] * T + param_gass[3])       #[m^3]
```

De tilpassede funksjonene vi har laget får med seg formen til de eksperimentelle verdiene, men vi ser at det er noen avvik.  Vi ser at for væskevolum så blir ikke tilpassingen like god som med gassvolum og latent varme. Årsaken til dette er at funksjonen vi ønsker å tilpasse ikke er god nok for å representere disse verdiene.

Videre så ønsker vi å bruke disse tilpassede funksjonene til å se damptrykk-kurven til vann. Dette gjør vi ved å bruke Simpson's metode på uttrykket for den deriverte av damptrykk-kurven (Clapeyrons ligning):

$$ \frac{dp}{dT} =\frac{\Delta S}{\Delta V} =  \frac{L(T)}{T\cdot(V_{gass}(T) - V_{væske}(T))}.$$
 


```python
#2c)
def simpson(func,a,b,N): 
    """
    Funksjon som finner arealet under en graf med Simpson's metode
    input:
        func: funksjon som skal integreres
        a: startverdi for integrasjonen 
        b: sluttverdi for integrasjonen
        N: antall delintervaller
    output:
        sum: en array med verdier for integralet fra a til b 
    """
    h = (b-a)/N                               #steglengde 
    x = np.linspace(a,b, N)                   #x-akse
    sum = np.zeros(N)                         #selve summen som en array, gjør det enklere å plotte 
    sum[0] = func(x[0])                       #startverdi for summen 
    for i in range(1,N):                      #for-løkke som beregne summen i hvert punkt 
        if(i%2==0 ):                          #partall sum 
            sum[i] = sum[i-1] + 2*func(x[i])
        else:                                 #oddetall sum 
            sum[i] = sum[i-1] + 4*func(x[i])
    sum[-1]= sum[-2] + func(x[-1])            #siste sum
    sum *= h/3                                #forholdstall for summen
    return sum
```

```python
def dp(T):
    """
    Funksjon for den deriverte av trykket med hensyn på temperaturen med tilpassede funksjoner
    input:
        T: temperaturen som den deriverte skal returneres for 
    output:
        Den deriverte
    """
    return latent_varme_func(T)/(T *(gass_volum_func(T) - væske_volum_func(T))) #Pa/K 
```

```python
#Bruker Simpsons metode på Clapeyrons ligning, for numerisk anslag av damptrykkurven
T1 = np.linspace(274, 647, 374)
p = simpson(dp, T1[0], T1[-1],len(T1))

#Plotter
plt.subplots(1,2, figsize = (14,5))

#Plotter for tilpasningsfunksjonene
plt.subplot(1,2,1)
plt.title(f"Tilpasset damptrykkurve med Simpson's metode")
plt.plot(T1, p/10**5, label = f'Tilpasset trykk')
plt.plot(kelvin_array, pressure_array/10**5, 'ro')
plt.plot(kelvin_array, pressure_array/10**5, 'r-', label = f'Eksperimentelt trykk')
plt.xlabel(f'Temperatur[K]')
plt.ylabel(f'Trykk [bar]')
plt.legend()

plt.subplot(1,2,2)
plt.title(f'Clapeyrons ligning tilpasset')
plt.plot(T1, dp(T1)/10**5, label = r'$\frac{dp}{dT}$')
plt.xlabel(f'Temperatur[K]')
plt.ylabel(r'$dp/dT$ [bar/K]')
plt.legend()
plt.show()

#Printer ut kritiske verdier for volum og latent varme
print(f"Væskevolum ved kritisk temperatur: {væske_volum_func(T1[-1])*10**6:.3f} ml")
print(f"Gassvolum ved kritisk temperatur: {gass_volum_func(T1[-1])*10**6:.3f} ml")
print(f"Latent varme ved kritisk temperatur: {latent_varme_func(T1[-1])/10**3:.3f} kJ")
print(f"Clapeyrons ved kritisk temperatur: {dp(T1[-1])/10**5:.3f} bar/K")
```

Fra plottet så ser vi at det numerisk beregnede trykket ligger under de eksperimentelle verdiene i starten, men når vi nærmer oss den kritiske tempreraturen så går det numeriske trykket over de eksperimentelle verdiene. Dette skyldes nok tilpasningskurven for væskevolumet. Et annet problem som dukker opp er fortegnet i Clapeyrons ligning. Den latente varmen vil bli negativ ved kritisk temperatur, slik at Clapeyrons ligning blir negativ. Dette er urealistisk! I tillegg ser man at tilpasningsfunksjonene ikke gir oss likt væskevolum og gassvolum ved kritisk trykk, noe som skulle vært tilfelle. Dermed kan man konkludere med at kurvetilpasningen fungerer bra for normale temperaturer, men svikter når vi nærmer oss den kritiske temperaturen.

Til nå har man sett at kurvetilpasning fungerer godt, men har svakheter når man nærmer seg kritisk temperatur. En annen ulempe er at man bruker selvskrevne funksjoner, som sjeldent passer godt til de numeriske verdier. Dermed skal man nå teste en ny metode, interpolasjon. Dette skal man utføre ved bruk av kubiske spliner.

```python
#2d)
#funksjon som interpolerer punktene i temp_volum og volum_væske
interpolert_væske_func = interpolate.CubicSpline(temp_volum, volum_væske) 

#funksjon som interpolerer punktene i temp_volum og volum_gass
interpolert_gass_func = interpolate.CubicSpline(temp_volum, volum_gass) 

#funksjon som interpolerer punktene i temp_latent_varme og latent_varme
interpolert_latent_varme_func = interpolate.CubicSpline(temp_latent_varme, latent_varme) 

def dp_cs(T):
    """
    Funksjon for den deriverte av trykket med hensyn på temperaturen, men med interpolerte funksjoner 
    beregnet ut ifra kubiske spliner
    input:
        T: temperaturen som den deriverte skal returneres for 
    output:
        Den deriverte 
    """
    return interpolert_latent_varme_func(T) / (T*(interpolert_gass_func(T) - interpolert_væske_func(T))) #Pa/K
```

```python
#Lager arrayer med flere punkter for å få bedre graf for både temp_volum og temp_latent_varme
T2 = np.linspace(temp_volum[0], temp_volum[-1], 1000)

T3 = np.linspace(temp_latent_varme[0], temp_latent_varme[-1], 1000)
#Plotter
plt.subplots(1,2, figsize = (10,5))
plt.subplot(1,2,1)
plt.title(f'Interpolert og tilpasset funksjon for væskevolum')
plt.plot(T2, interpolert_væske_func(T2)*10**6, 'blue',label ='Interpolert funksjon')
plt.plot(T2, væske_volum_func(T2)*10**6, "r",label="Tilpasset funksjon")
plt.plot(temp_volum, volum_væske*10**6, '.b', label =  f'Eksperimentelle verdier')
plt.xlabel(f'Temperatur [K]')
plt.ylabel(f'Volum [ml]')
plt.legend()

plt.subplot(1,2,2)
plt.title(f'Interpolert og tilpasset funksjon for gassVolum')
plt.plot(T2, interpolert_gass_func(T2), 'blue', label ='Interpolert funksjon')
plt.plot(T2, gass_volum_func(T2), "r",label="Tilpasset funksjon")
plt.plot(temp_volum, volum_gass, '.b', label =  f'Eksperimentelle verdier')
plt.xlabel(f'Temperatur [K]')
plt.ylabel(fr'Volum gass [$m^3$]')
plt.legend()
plt.show()


plt.title(f'Interpolert og tilpasset funksjon for latent varme')
plt.plot(T3, interpolert_latent_varme_func(T3)/10**3, "blue", label ='Interpolert funksjon')
plt.plot(T3, latent_varme_func(T3)/10**3,"r",label="Tilpasset funksjon")
plt.plot(temp_latent_varme, latent_varme/10**3, ".b",label =  f'Eksperimentelle verdier')
plt.xlabel(f'Temperatur [K]')
plt.ylabel(f'Latent varme [kJ]')
plt.legend()
plt.show()
```

Vi ser fra plottene at funksjonene for interpolasjon og tilpasningene har lik form. Begge metodene samsvarer med de eksperimentelle verdiene. En tydelig forskjell mellom metodene er at de interpolerte funksjonene "svinger" mer enn tilpasningsfunksjonene gjør. Det skyldes at de interpolerte funksjonene tvinges gjennom alle punktene og kan da føre til at vi får svinginger mellom punkter. Tilpasningsfunksjonene vil ha en glattere og mer forutsigbar kurve, men på bekostning av at noen eksperimentelle verdier ikke ligger på funksjonen. En annen måte å forbedre tilpasningsfunksjonen ville vært å ha flere parametere, men også da mer kompliserte uttrykk. 

For latent varme så er begge funksjonene veldig sterke, men svingningene til den interpolerte funksjonen er veldig tydelige. Så for latent varme vil kurvetilpasning være den som passer best. For væskevolumet er den interpolerte funksjonen mye bedre, men får noen unaturlige svinginger ved høyere temperatur. For gassvolumet fungerer både interpolasjon og kurvetilpasning godt.


Videre skal vi bruke Simpson's metode, Clapeyrons ligning og interpolasjon til å beregne damptrykkurven.

```python
#Beregner trykket med de interpolerte funksjonene og simpsonsmetode 
pressure_interpolation = simpson(dp_cs, T2[0],T2[-1],len(T2)) 

plt.subplots(1,2, figsize = (14,5))

plt.subplot(1,2,1)
plt.title(f'Trykket som funksjon av temperaturen')
plt.plot(T2[:-1], pressure_interpolation[:-1]/10**5, label = 'Trykk med interpolasjon og Simpson') 
plt.plot(kelvin_array, pressure_array/10**5,'.r', label = f'Eksperimentelle verdier')
plt.xlabel(f'Temperatur [K]')
plt.ylabel(f'Trykk [bar]')
plt.legend()

plt.subplot(1,2,2)
plt.title(f'Clapeyrons ligning interpolert ')
plt.plot(T2[:-1], dp_cs(T2[:-1])/10**5, label = r'$\frac{dp}{dT}$')
plt.xlabel(f'Temperatur[K]')
plt.ylabel(fr'Trykk [bar/K]')
plt.legend()
plt.show()

print(f'Volum til gass ved kritisk temperatur = {interpolert_gass_func(T2[-1])*10**6:.3f} ml')
print(f'Volum til væske ved kritisk temperatur = {interpolert_væske_func(T2[-1])*10**6:.3f} mml')
print(f'Latent varme ved kritisk temperatur = {interpolert_latent_varme_func(temp_latent_varme[-1]):.3e} J')
print(f'Clapeyrons ligning ved kritisk temperatur = {dp_cs(T2[-1]):.3e} Pa/K')
print(f'Trykk ved kritisk temperatur = {pressure_interpolation[-1]:.3e} Pa')
```

Fra plottet observerer man at trykket beregnet fra de interpolerte funksjonene representeter de eksperimentelle dataene godt. Men man ser at kurven ikke når helt opp til kritisk temperatur. Her er volumet til gassen og væsken likt og vi får da deling på $0$ i Clapeyrons ligning. Delingen på $0$ ville vanligvis fått problemer for numerikken, men vi unngår dette ved å ikke plotte siste verdi av trykket og Clapeyrons ligning. En annen feil oppstår for den kritiske temperaturen. Her skal den latente varmen bli $0$, men siden de interpolerte funksjonene ikke er trent på samme temperaturverdier, blir den latente varmen negativ. 

Fra plottet av damptrykkurven ser man at avviket ikke er stort før den kritiske temperaturen. Her er det viktig at vi har interpolert for samme temparaturverdier. Dette skjer ikke, siden temperaturmålingene er hentet fra to forskjellige forsøk. Til å konkludere med er interpolasjon en god metode for å plotte damptrykket, men blir ugyldig når vi nærmer oss den kritiske temperaturen. For å rette opp i dette burde man ha gjort målinger for gassvolum, væskevolum og latent varme for like temperaturverdier.


Videre så ønsker vi å sammenlikne damptrykk-kurven for interpolasjon, tilpasningsfunksjonene, Van der Waals tilstandsfunksjon og idell gasslov.

```python
def trykk_ideell_gass(T, L):
    """
    trykk for ideell gass lov 
    input:
        T: temperatur-array
        L: konstant latent varme 
    output:
        trykket som funksjon av temperaturen
    """
    return pressure_array[0]*np.exp(L/R*(1/T[0] -1/T)) #Bruker eksperimentell verdi for p0, 

param_ideell_gass, param_cov_ideell_gass = scipy.optimize.curve_fit(trykk_ideell_gass, kelvin_array, pressure_array)
```

Fra plottet observerer man at Van der Waals tilstandsligning fungerer mye dårligere enn alle de andre metodene. Ideell gasslov med konstant latent varme $L$, ligger litt under de eksperimentelle målingene i hele intervallet, men kurven passer godt. Dette er en overraskende observasjon da den latente varmen endres mye for høyere temperaturer. En annen ovveraskende observasjon er at ideell gasslov er en bedre tilnærming til damptrykkurven enn Van der Waals tilstandsligning, da Van der Waals tilstandsligning har et mer reelt uttrykk for trykket. Dette ville nok endret seg om man plottet for mange mol $H_2O$, da volumet optatt av molekylene og intermolekylære krefter ville hatt en større betydning. 

Man observerer likevel at det er interpolasjonen som passer kurven best. Interpolasjonen går gjennom nesten alle de eksperimentelle målingene, og uten unaturlige svingninger. Derimot kurvetilpasningen representerer grafen godt men bommer på mange punkter ved høyere temperaturer. 

Det er verdt å observere at alle metodene, bortsett fra Van der Waals tilstandsligning, får utfordringer med å representere trykket ved kritisk temperatur.
